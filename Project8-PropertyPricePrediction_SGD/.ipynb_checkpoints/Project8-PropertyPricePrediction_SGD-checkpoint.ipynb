### Problem Statement

We are going to build a linear regression model with stochastic gradient descent to predict the price of the property from the dataset having attributes such as sale type, sale condition etc. 

### Data Preprocessing

We will check the data distribution of variables and perform transformation if a variableâ€™s distribution is skewed. We will perform label encoding on categorical variables.

### Model Comparison

We will build linear regression models and compare results.

### Model Selection

We will select the best model. Our model selection to be based on model accuracy.

### Expected Outcome

Low RMSE and high coefficient of determination(R^2) is expected while predicting the outcome using test data.

# Missing Values

Most libraries will give you an error if you try to build a model using data with missing values. Python libraries represent 
missing number as nan which is sort for "not a number". You can detect which cells have missing values, and then count 
how many there are in each column with the command isnull().sum()

```python
#Check for missing values
missing = alldata.isnull().sum()
missing
```

```python
#check for missing values
missing = alldata.isnull().sum()
missing = missing[missing > 0]

#we can use bar plot to check missing values
missing.sort_values(inplace=True)
missing.plot.bar()
plt.show()
```

There are 35 attributes having missing values.

# Missing value treatment for numerical features

```python
#fill na with median
alldata.fillna(alldata.median(), inplace = True) #only numerical variables are going to be imputed
```

```python
# Again check for missing values
missing = alldata.isnull().sum()
missing = missing[missing > 0]
#we can use bar plot to check missing values
missing.sort_values(inplace =True)
missing.plot.bar()
plt.show()
```

We can only replace missing values by mean whose data type is numeric. For a few columns there is lots of NaN entries.
However, reading the data description we find this is not missing data. For Pool_Quality, NaN is not missing data but means no pool, likewise for Fence_Qualiy, Fireplace_Quality etc.

# Missing value treatment for categorical features

```python
#Replace nan with None in these columns
cols_fillna = ['Pool_Quality','Miscellaneous_Feature','Lane_Type','Fence_Quality','Brick_Veneer_Type','Fireplace_Quality',
               'Garage_Quality','Garage_Condition', 'Electrical_System','Garage_Finish_Year', 
               'Exposure_Level','Basement_Condition','Basement_Height','BsmtFinType1','BsmtFinType2','Garage','Zoning_Class','Functional_Rate',
               'Utility_Type','Exterior1st','Sale_Type','Exterior2nd','Kitchen_Quality','Electrical_System']

for col in cols_fillna:
    alldata[col].fillna('None',inplace=True)
```

```python
#Import library for VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor

def cal_vif(x):
    
    #Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = x.columns
    vif["VIF"] = [variance_inflation_factor(x.values, i) for  i in range(x.shape[1])]
    
    return(vif)
```

```python
cal_vif(df_numeric_features)
```

# Scale our numeric variables using min-max normalization

```python
#apply the min-max scaling to our numeric variables
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()

#scaling down the numeric variables
df_housingdata_numcols = pd.DataFrame(min_max.fit_transform(df_numeric_features.iloc[:,0:38]),
columns = df_numeric_features.iloc[:,0:38].columns.tolist())                                      
```

```python
df_housingdata_numcols.head()
```

# Perform one-hot encoding on categorical variables

```python
for col in df_categorical_features.columns.values:
    one_hot_encoded_variables = pd.get_dummies(df_categorical_features[col], prefix=col)
    df_categorical_features = pd.concat([df_categorical_features,one_hot_encoded_variables], axis = 1)
    df_categorical_features.drop([col], axis = 1, inplace = True)
```

# Concatenate the numeric and encoded variables to the dataframe

```python
df_housedata = pd.concat([df_housingdata_numcols, df_categorical_features], axis = 1)
```

```python
df_housedata.columns
```

# Creating training and testing dataset using the train_test_split

```python
#Create feature and response variable set
#We create train and test sample from our dataset

from sklearn.model_selection import train_test_split

#create feature and response variables
X = df_housedata.drop(['Sale_Price'], axis = 1)
Y = df_housedata["Sale_Price"]

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

# Use SGDRegressor() to build a linear model

```python
from sklearn.linear_model import SGDRegressor 

lin_model = SGDRegressor()

#We fit our model with train data
lin_model.fit(x_train, y_train)

# We use predict() to predict our values
lin_model_predictions = lin_model.predict(x_test)

# we check the coefficient of determination with score()
print(lin_model.score(x_test,y_test))
```

Coefficient of determination tells about the goodness of fit of a model.

```python
print(lin_model.intercept_)
```

```python
print(lin_model.coef_)
```

# Mean Square Error

```python
#We check the root mean square error (RMSE)
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, lin_model_predictions)
rmse = np.sqrt(mse)
print(rmse)
```

RMSE is 0.0566 means the total loss(squareroot of the difference between the actual value and the predicted value).
The smaller the RMSE value, the better the model.

# Create a model with grid search

Grid search is an approach to hyperparameter tuning that will methodically build and evaluate a model for each combination of algorithm parameters specified in a grid.

```python
#ignore the deprecation warning

import warnings 

warnings.filterwarnings("ignore", category = DeprecationWarning)
warnings.filterwarnings("ignore", category = FutureWarning)

from sklearn.model_selection import GridSearchCV

# Grid search - this will take about 1 minute.
param_grid = {
    'alpha': 10.0 ** -np.arange(1,7),
    'loss': ['squared_loss', 'huber', 'epsilon_insensitive'],
    'penalty': ['l2', 'l1', 'elasticnet'],
    'learning_rate': ['constant', 'optimal', 'invscaling'],
    
}
clf= GridSearchCV(lin_model, param_grid)
clf.fit(x_train,y_train)
print("Best Score:" +str(clf.best_score_))
```

```python
?GridSearchCV
```

```python
pred_test = clf.predict(x_test); pred_test
```

```python
alldata['Sale_Price'].min()
```

```python
Sale_Price_Range = alldata['Sale_Price'].max()-alldata['Sale_Price'].min()
Sale_Price_Range
```

# Print best parameter values

```python
print('Best C:',clf.best_estimator_.C) 
print('Best alpha:',clf.best_estimator_.alpha) 
print('Best n_iter:',clf.best_estimator_.n_iter_no_change)
print('Best tol:',clf.best_estimator_.tol) 
print('Best eta0:',clf.best_estimator_.eta0) 
print('Best learning rate:',clf.best_estimator_.learning_rate)
```
